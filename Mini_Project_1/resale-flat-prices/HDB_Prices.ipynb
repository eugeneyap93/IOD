```python
import pandas as pd
import glob
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

#define path--using this cell to loop through the csv files from path into one dataframe
path = r'C:\Users\Eugene\Documents\IOD-Data Science\Mini_Project_1\resale-flat-prices'

#glob module is used to retrieve file names matching a specific pattern
#create a list of all the files to be loaded
all_files = glob.glob(path + "/*.csv")

df = []

#loop through all the files name in all_files list and append it to a list
for filename in all_files:
    file = pd.read_csv(filename, index_col=None, header=0)
    df.append(file)

#use pd.concat to turn df into a dataframe
resale_flat = pd.concat(df, axis=0, ignore_index=True)
```

```python
#Create new column lease_category to catergorise remaining at point of sale into 6 groups
conditions = [
    (resale_flat['updated_remaining_lease'] <= 50),
    (resale_flat['updated_remaining_lease'] > 50) & (resale_flat['updated_remaining_lease'] <= 60),
    (resale_flat['updated_remaining_lease'] > 60) & (resale_flat['updated_remaining_lease'] <=70),
    (resale_flat['updated_remaining_lease'] > 70) & (resale_flat['updated_remaining_lease'] <=80),
    (resale_flat['updated_remaining_lease'] > 80) & (resale_flat['updated_remaining_lease'] <=90),
    (resale_flat['updated_remaining_lease'] > 90)
    ]
values = ['<50', '51-60', '61-70', '71-80', '81-90','>90']
resale_flat['lease_category'] = np.select(conditions, values)
```

```python
#Plot resale count by lease category
catplot = sns.catplot(x="flat_type", col='lease_category', col_wrap=2, data=resale_flat, kind='count',order = resale_flat['flat_type'].sort_values().unique(),
                height=4, aspect=2.5);
catplot.set_axis_labels("Flat Types", "Resale Count")
```

```python
#Create a new dataframe to find out the average resale price by town, flat type and year
resale_avg_price = resale_flat.groupby(['town', 'flat_type','year'], as_index = False)[['resale_price']].mean()
```

```python
#Create a new dataframe to find average price difference based town and flat type
#average price difference --> Price of Max Year minus Price of Min Year

resale = resale_avg_price[resale_avg_price['year'] >= '2010']

# Find unique values in resale_avg_price 
town_flat_type = resale[['town', 'flat_type']].drop_duplicates()

#Initialise a data frame
df = pd.DataFrame(columns=['town', 'flat_type', 'price_diff'])

#Loop through the values and append to dataframe df
for a, b in town_flat_type.itertuples(index=False):
    min_y = resale['year'][(resale['town'] == a) & (resale['flat_type'] == b)].min()
    max_y = resale['year'][(resale['town'] == a) & (resale['flat_type'] == b)].max()
    price_min_y = resale['resale_price'][(resale['town'] == a) & (resale['flat_type'] == b) & (resale['year'] == min_y)]
    price_min = price_min_y.iloc[0]
    price_max_y =  resale['resale_price'][(resale['town'] == a) & (resale['flat_type'] == b) & (resale['year'] == max_y)]
    price_max = price_max_y.iloc()[0]
    diff = price_max - price_min
    df = df.append({'town': a,'flat_type': b, 'price_diff':diff}, ignore_index = True)
```
